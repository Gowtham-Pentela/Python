```python
'''Write a short Python function, minmax(data), that takes a sequence of
one or more numbers, and returns the smallest and largest numbers, in the
form of a tuple of length two. Do not use the built-in functions min or
max in implementing your solution'''
def minmax(data):
    if not data:return
    minimum=maximum=data[0]
    for i in range(1,len(data)):
        if data[i] > maximum:
            maximum = data[i]
        elif data[i] < minimum:
            minimum=data[i]
    return (minimum, maximum)
print(minmax([5,-1,9,10]))
print(minmax([2]))
print(minmax([]))
```

```python
'''Write a short Python function that takes a sequence of integer values and
determines if there is a distinct pair of numbers in the sequence whose
product is odd.'''
def odd_distinct_pair(data):
    if not data:return False
    i=0
    j=len(data)-1
    while i < len(data)-1:
        if (data[i]* data[j])%2!=0:
            return True
        if j == i:
            i+=1
            j=len(data)-1
        else: 
            j-=1
    return False
print(odd_distinct_pair([2,4,6,5,9]))
```

```python
'''In Section 1.8, we provided three different implementations of a generator
that computes factors of a given integer. The third of those implementations, from page 41, was the most efficient, but we noted that it did not
yield the factors in increasing order. Modify the generator so that it reports
factors in increasing order, while maintaining its general performance advantages.'''
def factors(n):
    """Generate factors of n in increasing order."""
    small_factors = []
    for i in range(1, int(n**0.5) + 1):  
        if n % i == 0:  
            yield i  
            if i != n // i:  
                small_factors.append(n // i)  
    for factor in reversed(small_factors):
        yield factor


print(list(factors(100)))  
print(list(factors(36)))  
```

```python
'''The p-norm of a vector v = (v1,v2,...,vn) in n-dimensional space is defined as
||v||
For the special case of p = 2, this results in the traditional Euclidean
norm, which represents the length of the vector. For example, the Euclidean norm of a two-dimensional vector with coordinates (4,3) has a
Euclidean norm of √
42 +32 = √16+9 = √
25 = 5. Give an implementation of a function named norm such that norm(v, p) returns the p-norm
value of v and norm(v) returns the Euclidean norm of v. You may assume
that v is a list of numbers.'''
def p_norm(v):
    n=len(v)
    total=0
    for i in v:
        total += i**n
    return total**(1/n)
v=(4,3)
print(p_norm(v))
```

```python
'''
Write a Python program that can take a positive integer greater than 2 as
input and write out the number of times one must repeatedly divide this
number by 2 before getting a value less than 2.
'''
def count_2s(n):
    count=0
    if n<2:
        return None
    while n>=2:
        n=n//2
        count+=1
    return count
print(count_2s(2))
print(count_2s(10))
```

```python
'''Write a Python program that simulates a handheld calculator. Your program should process input from the Python console representing buttons
that are “pushed,” and then output the contents of the screen after each operation is performed. Minimally, your calculator should be able to process
the basic arithmetic operations and a reset/clear operation'''
def handheld_calculator():
    print("Handheld Calculator: Enter numbers and operators one by one.")
    print("Type '=' to display the result, 'C' to clear, and 'exit' to quit.\n")

    expression = ""  # Stores the current input expression
    
    while True:
        user_input = input("Press button: ").strip()

        if user_input.lower() == "q":  # Quit the calculator
            print("Exiting calculator. Goodbye!")
            break
        elif user_input.upper() == "C":  # Clear the screen
            expression = ""
            print("Calculator reset.")
        elif user_input == "=":  # Evaluate the expression
            try:
                result = eval(expression)  # Safely evaluate expression
                print(f"Screen: {result}")
                expression = str(result)  # Store result for further calculations
            except Exception as e:
                print("Error:", e)
                expression = ""  # Reset on error
        else:
            try:
                # Allow numbers, decimal points, and operators
                if user_input.replace('.', '', 1).isdigit() or user_input in "+-*/":
                    expression += user_input  # Append input to expression
                    print(f"Screen: {expression}")
                else:
                    print("Invalid input. Use numbers, operators, '=', 'C', or 'exit'.")
            except Exception as e:
                print("Error:", e)

# Call the function
handheld_calculator()
```

```python
'''The birthday paradox says that the probability that two people in a room
will have the same birthday is more than half, provided n, the number of
people in the room, is more than 23. This property is not really a paradox,
but many people find it surprising. Design a Python program that can test
this paradox by a series of experiments on randomly generated birthdays,
which test this paradox for n = 5,10,15,20,...,100.'''
import random

def has_duplicate_birthday(num_people):
    """Returns True if at least two people in the group share a birthday."""
    birthdays = [random.randint(1, 365) for _ in range(num_people)]
    return len(birthdays) != len(set(birthdays))  

def birthday_paradox_simulation(num_people, trials=10000):
    """Runs multiple trials and computes the probability of a shared birthday."""
    count_matches = sum(has_duplicate_birthday(num_people) for _ in range(trials))
    return count_matches / trials  

print("n\tProbability of Shared Birthday")
print("-" * 40)
for n in range(5, 101, 5):
    probability = birthday_paradox_simulation(n)
    print(f"{n}\t{probability:.4f}")
```

```python
'''Write a Python program that inputs a list of words, separated by whitespace, and outputs how many times each word appears in the list. You
need not worry about efficiency at this point, however, as this topic is
something that will be addressed later in this book.'''
def word_count():
    text = input("Enter a list of words separated by spaces:\n")
    words = text.split()  
    word_freq = {}  

    for word in words:
        word_freq[word] = word_freq.get(word, 0) + 1  

    print("\nWord Frequencies:")
    for word, count in word_freq.items():
        print(f"{word}: {count}")


word_count()
```
